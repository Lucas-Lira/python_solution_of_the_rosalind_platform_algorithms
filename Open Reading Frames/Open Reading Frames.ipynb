```python
def complementing_a_strand_of_DNA(content):
    result = ''
    for i in range(len(content)-1, -1, -1):
        if content[i] == 'A':
            result += 'T'
        if content[i] == 'T':
            result += 'A'
        if content[i] == 'C':
            result += 'G'
        if content[i] == 'G':
            result += 'C'
    return result
```
